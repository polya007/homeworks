```python
# Задача №1

n = int(input('Сколько пингвинов заказывали? Введите число от 1 до 9: '))
a = '    _~_    '  
b = '   (o o)   '
c = '  /  V  \  '
d = ' /(  _  )\\ '
e = '   ^^ ^^   '

print(n)
if n > 9: 
    print("Введено неверное число")
elif n < 1:
    print("Введено неверное число")
elif 1 < n < 9:
    print(a*n)
    print(b*n)
    print(c*n)
    print(d*n)
    print(e*n)
```
